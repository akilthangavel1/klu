# Data Preprocessing with Pandas

This notebook covers the standard steps for proper preprocessing of data using the `pandas` library.

```python
import pandas as pd
import numpy as np

# Let's create a mock dataset to demonstrate data preprocessing
data = {
    'Id': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'Age': [25, np.nan, 22, 29, 35, 29, np.nan, 42, 31, 28],
    'Salary': [50000, 54000, 50000, 60000, 75000, 60000, 45000, np.nan, 58000, 52000],
    'Country': ['France', 'Spain', 'Germany', 'Spain', 'France', 'Spain', 'Germany', 'France', 'Germany', 'France'],
    'Purchased': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)
df
```

## 1. Inspecting the Data

```python
# Check total records and columns
print("Shape of DataFrame:", df.shape)

# Display data types
print("\nData Types:")
print(df.dtypes)

# Overview of numerical columns
df.describe()
```

## 2. Handling Missing Data (Imputation)

```python
# Check for missing values
print("Missing values per column:\n", df.isnull().sum())

# Replace missing 'Age' with the mean age
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

# Replace missing 'Salary' with the median salary
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)

print("\nMissing values after imputation:\n", df.isnull().sum())
df
```

## 3. Handling Duplicate Records

```python
# Checking for duplicate rows based on all columns
duplicates = df.duplicated()
print(f"Found {duplicates.sum()} duplicate rows.")

# Drop duplicate rows
df = df.drop_duplicates()
print("Shape after dropping duplicates:", df.shape)
df
```

## 4. Dropping Unnecessary Columns

```python
# The 'Id' column does not provide useful information for most machine learning models
if 'Id' in df.columns:
    df = df.drop('Id', axis=1)
df
```

## 5. Encoding Categorical Variables

```python
# One-Hot Encoding for 'Country' column (Nominal Data)
df = pd.get_dummies(df, columns=['Country'], prefix='Country', drop_first=True)

# Binary Mapping for 'Purchased' column
df['Purchased'] = df['Purchased'].map({'No': 0, 'Yes': 1})
df
```

## 6. Feature Scaling

```python
# Standardizing the numerical columns (Age and Salary) so they have a mean of 0 and a standard deviation of 1
df['Age'] = (df['Age'] - df['Age'].mean()) / df['Age'].std()
df['Salary'] = (df['Salary'] - df['Salary'].mean()) / df['Salary'].std()

# The dataset is now fully preprocessed and ready for machine learning algorithms!
df
```
